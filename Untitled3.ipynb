Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and processing

```python
#loading the csv data to a pandas dtaframe
heart_data = pd.read_csv('/content/Heart_Disease_Prediction.csv')
```

```python
#print first 5 rows of dataset
heart_data.head()
```

```python
#print last 5 rows of data
heart_data.tail()
```

```python
#number of rows and columns in the dataset
heart_data.shape
```

```python
#getting some info. about dataset
heart_data.info()
```

```python
#checking for missing values
heart_data.isnull().sum()
```

```python
#statistical measure of the data
heart_data.describe()
```

```python
#checking the distribution of target variable
if 'Heart Disease' in heart_data.columns:
    # If yes, proceed as before, but replace values
    counts = heart_data['Heart Disease'].value_counts()
    counts = counts.rename({1: 'presence', 0: 'absence'})
    print(counts)
else:
    # If not, print an error message or try a different column name
    print("Error: 'Heart Disease' column not found in DataFrame.")
    # You can also try to find the correct column name by printing all columns:
    print("Available columns:", heart_data.columns)
```

Absence-->defective heart
Present-->healthy heart

Splitting features and Heart Disease columns

```python
X = heart_data.drop(columns='Heart Disease', axis=1)
Y = heart_data['Heart Disease']
Y
```

splitting the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training

Logistic regression

```python
model=LogisticRegression(max_iter=10000)
```

```python
#Training the Logistic regression model with training data

model.fit(X_train, Y_train)
```

model evaluation

Acuracy score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```

Building a predictive system

```python
input_data = (64,1,4,128,263,0,0,105,1,0.2,2,1,7)

input_data_df = pd.DataFrame([input_data], columns=X.columns)

# 2. Use the DataFrame for prediction
Heart_Disease_Prediction = model.predict(input_data_df)


if (Heart_Disease_Prediction[0]== 'Absence'):
  print('The Person does not have Heart Disease')
else:
  print('The Person have Heart Disease')
```
